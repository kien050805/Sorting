```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {"array_length": [10, 100, 1000, 10000, 100000],
        "insertion_sort": [0, 5, 387, 34551, 3436440],
        "merge_sort": [0, 8, 112, 1382, 16374],
        "quick_sort": [0, 5, 83, 1118, 13514],
        "improved_quick_sort": [0, 6, 85, 1123, 13379],
        "randomized_quick_sort": [53, 568, 5795, 57818, 579853],
        "cpp_sort": [0, 1, 24, 335, 4115]}

df = pd.DataFrame(data)
df = df.set_index("array_length")
df.head(10)
```

```python
sns.set(style="darkgrid")
tips = sns.load_dataset("tips")
color = sns.color_palette()[2]
f, ax = plt.subplots(figsize=(7, 7))
ax.set(xscale="log", yscale="log")

plot = sns.lineplot(data=df)
plot.set_xlabel("Array length")
plot.set_ylabel("Runtime in microseconds in log scale")
plot.set_title("Runtime average of each sorting algorithm")
```
